```python
import cv2
import sys
from matplotlib import pyplot as plt
```

```python
pathImage=r'image.JPG'
cascadeFile=r"D:\IA\EXERCICES\python\IA\detection_de_visages\opencv-3.4\data\haarcascades\haarcascade_frontalface_alt.xml"
classCascade=cv2.CascadeClassifier(cv2.data.haarcascades + cascadeFile)
image=cv2.imread(pathImage)
imagegray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
#charger le classifieur
classCascade.load(cascadeFile)
```

```python
#detection de visages
faces=classCascade.detectMultiScale(imagegray, 1.1, 5, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)

#rectangles autours des visages detectés 
for (x,y,w,h) in faces :
    cv2.rectangle(image, (x,y),(x+w,y+h), (255,0,0),2)
plt.imshow(image)    
```
